## Evaluation

```python
model_evaluator = model_evaluation_fc(config_dic)
```

```python
#read image and evaluate
ix = 0
name_file = pd_data.loc[ix]['input_file']
name_dir = pd_data.loc[ix]['input_dir']
prefix = name_file.split('.')[0]

raw_image = np.load(name_dir + name_file)
raw_image = raw_image['image']

mask_softmax = model_evaluator(raw_image)[0,...]
```

```python
ix = 0
plt.figure(figsize=(10,5))
for i in np.arange(3):
    plt.subplot(1,3,i+1)
    plt.imshow(mask_softmax[...,ix])
    ix += 1
plt.show()

plt.figure(figsize=(10,5))
for i in np.arange(2):
    plt.subplot(1,2,i+1)
    plt.imshow(mask_softmax[...,ix])
    ix += 1
plt.show()
```

```python
th = 0.8
instance_fore = np.array(mask_softmax[...,-2])
instance_fore[instance_fore>th*np.max(instance_fore)] = 1
instance_fore[instance_fore<=th*np.max(instance_fore)] = 0

instance_fore_aux = morphology.dilation(instance_fore,selem = morphology.disk(2))
outline = instance_fore_aux-instance_fore

out_image = np.zeros([raw_image.shape[0],raw_image.shape[1],3])
out_image[...,1] = raw_image
out_image[...,0] = outline

plt.figure(figsize=(8,8))
plt.imshow(out_image)
plt.show()
```
